```python
import pandas as pd
import numpy as np
```

```python
dou_old = pd.read_excel('data/dou_old.xlsx', sheet_name = 'Planilha Exec Legisl revisada')
dou_new = pd.read_excel('data/dou_new.xlsx')
```

```python
dou_old
dou_old.columns = dou_old.iloc[0].values #makes the columns (first row) useful queriable
dou_old = dou_old.drop(dou_old.index[0]) #drop the first row
dou_old
```

```python
dou_new
```

```python
dou_new.replace('IDEM', method = 'ffill', inplace = True)
```

```python
dou_new['Idx'] = dou_new['Idx'].fillna(method = 'ffill')
dou_new['title'] = dou_new['title'].fillna(method = 'ffill')
dou_new['date'] = dou_new['date'].fillna(method = 'ffill')
dou_new['type'] = dou_new['type'].fillna(method = 'ffill')
dou_new['publisher'] = dou_new['publisher'].fillna(method = 'ffill')
dou_new['origin'] = dou_new['publisher'].fillna(method = 'ffill')
dou_new['Incidence of WHO mentions'] = dou_new['Incidence of WHO mentions'].fillna(method = 'ffill')
```

```python
dou_new
```

```python
set(dou_new['TYPE OF MENTION'])
```

```python
imprecise = []
generic = []
precise = []

for mention_type in list(dou_new['TYPE OF MENTION']):
    if type(mention_type) == str:
        if 'Impre' in mention_type:
            imprecise.append('Yes')
            generic.append('No')
            precise.append('No')
        elif 'Prec' in  mention_type:
            imprecise.append('No')
            generic.append('No')
            precise.append('Yes')
        elif 'Gen' in  mention_type:
            imprecise.append('No')
            generic.append('No')
            precise.append('Yes')
    else:
        imprecise.append(np.nan)
        generic.append(np.nan)
        precise.append(np.nan)
```

```python
print(len(imprecise), len(precise), len(generic))
```

```python
dou_new['mencao_imprecisa_a_doc'] =  imprecise
dou_new['mencao_especifica_a_doc'] = precise
dou_new['mencao_oms_geral'] = generic
```

```python
set(dou_new['NAME OF MENTIONED WHO DOCUMENT (IF ANY)'])
```

```python
soft_hard = []
for doc in list(dou_new['NAME OF MENTIONED WHO DOCUMENT (IF ANY)']):
    if type(doc) == str:
        if 'Tobacco' in doc:
            soft_hard.append('Hard')
        elif 'Tabaco ' in doc:
            soft_hard.append('Hard')
        elif 'Classification of Diseases' in doc:
            soft_hard.append('Hard')
        else:
            soft_hard.append('Soft')
    else:
        soft_hard.append(np.nan)
```

```python
dou_new['hard_soft'] = soft_hard
```

```python
print(dou_old.columns)
print(dou_new.columns)
```

```python
idxs = list(dou_old['idx']) + list(dou_new['Idx'])
title = list(dou_old['title']) + list(dou_new['title'])
date = list(dou_old['date']) + list(dou_new['date'])
typ =  list(dou_old['type']) + list(dou_new['type'])
origin = list(dou_old['origin']) + list(dou_new['origin'])
section =  list(dou_old['section']) + list(dou_new['section'])
subject_theme = list(dou_old['subject_theme']) + list(dou_new['General subject-matter (BR)'])
health_issue = list(dou_old['major_health_issue']) + list(dou_new['HEALTH TOPIC'])
covid = list(dou_old['mencao_covid']) + list(dou_new['Mentions COVID?'])
doc_cited = list(dou_old['doc_citado']) + list(dou_new['NAME OF MENTIONED WHO DOCUMENT (IF ANY)'])
direct_mention = list(dou_old['forma_mencao']) + list(dou_new['DIRECT MENTION: YES OR NO'])
precise_mention = list(dou_old['mencao_especifica_a_doc']) + list(dou_new['mencao_especifica_a_doc'])
imprecise_mention = list(dou_old['mencao_imprecisa_a_doc']) + list(dou_new['mencao_imprecisa_a_doc'])
generic_mention = list(dou_old['mencao_oms_geral']) + list(dou_new['mencao_oms_geral'])
hard_soft = list(dou_old['hard_soft']) + list(dou_new['hard_soft'])
```

```python
dou = pd.DataFrame(data = {'idxs': idxs, 'title': title, 'date': date, 'typ': typ, 'origin': origin, 
                           'section': section,
                          'subject_theme': subject_theme, 'health_issue': health_issue, 'covid':covid,
                          'doc_cited': doc_cited, 'direct_mention': direct_mention, 
                           'precise_mention': precise_mention, 'imprecise_mention': imprecise_mention,
                          'generic_mention': generic_mention, 'hard_soft': hard_soft})
```

```python
dou.to_csv('dou.csv', index = False)
dou.to_excel('dou.xlsx', index = False)
```
